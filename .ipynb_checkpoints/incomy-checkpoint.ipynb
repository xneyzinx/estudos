```python



df[df['income']=="<=50K"].groupby('education')['age'].count().plot(kind='bar', figsize=(16,12), fontsize=20, color='b')
df[df['income']==">50K"].groupby('education')['age'].count().plot(kind='bar', figsize=(16,12), fontsize=20, color='r' )


```

```python
df2 = df[df['income']==">50K"].groupby('education')['income'].count()
df3 = df[df['income']=="<=50K"].groupby('education')['income'].count()


d = pd.concat([df2, df3], axis=1)
d.plot(kind='bar', stacked=True, figsize=(16,12), fontsize=20, grid=True, legend = False)
```
