```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:\\Users\\ney.dantas\\Desktop\\Cursos\\DataAnalyse\\Casas\\kc_house_data.csv")

df.describe()
```

```python
df["date"] = pd.to_datetime(df.date)
```

```python
df.dtypes
```

```python
df["sqft_living"] = df["sqft_living"]*0.092903
df.rename(columns={"sqft_living":"sqm_living"}, inplace=True)

```

```python
df.dtypes
```

```python
df["sqft_lot"] = df["sqft_lot"]*0.092903
df.rename(columns={"sqft_lot":"sqm_lot"}, inplace=True)

df["sqft_above"] = df["sqft_above"]*0.092903
df.rename(columns={"sqft_above":"sqm_above"}, inplace=True)

df["sqft_basement"] = df["sqft_basement"]*0.092903
df.rename(columns={"sqft_basement":"sqm_basement"}, inplace=True)

df["sqft_living15"] = df["sqft_living15"]*0.092903
df.rename(columns={"sqft_living15":"sqm_living15"}, inplace=True)

df["sqft_lot15"] = df["sqft_lot15"]*0.092903
df.rename(columns={"sqft_lot15":"sqm_lot15"}, inplace=True)

df.dtypes
```

```python
df.price = df.price/1000
df.head()
```

```python
#POUCAS CASAS CUSTAM MENOS DE 100.000 DOLARES. A MAIORIA CUSTA MENOS QUE 1 MILHÃO DE DÓLARES
np.log10(df.price).hist(bins=100, figsize=(16,8)).set(xticks=range(5), xticklabels=["0k", "10k", "100k", "1000k", "10000k"],
                                                      xlabel="House Price", ylabel="Number of Houses")
```
