```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:\\Users\\ney.dantas\\Desktop\\Cursos\\DataAnalyse\\San Francisco\\Library_Usage.csv")

df.head()

```

```python
df = df.iloc[: , [1,2,3,4,6,7,8,10,11,12,13,14]]
df.head()
```

```python

df.columns = ["patron_type", "total_checkouts", "total_renewals", "age_range", "home_library", "active_month",
             "active_year", "notice_preference", "provided_email", "year_patron_registered", "outside_county",
             "supervisor_district"]


df["year_patron_registered"].value_counts()
```

```python
def convert_years(year):
    if year == "None":
        return year
    else:
        return int(year)
    

df["active_year"] = df["active_year"].apply(convert_years)

```

```python
patron_registration = df["year_patron_registered"].value_counts().sort_index().drop(2003)
patron_registration.iloc[:]
```

```python
fig, ax = plt.subplots(1,1, figsize=(16, 8))
fig.subplots_adjust(hspace=0.4)

plot_1 = patron_registration.plot(marker="o", xticks=range(2004, 2017), xlim=[2003.5, 2016.5])
plot_1.set_title("Growth of Library Users", fontweight="bold")
plot_1.set(xlabel="Year", ylabel="Number of People that registered")
plt.grid()
plt.show()
```

```python
total_checkouts = df["total_checkouts"].value_counts().sort_index()
total_checkouts.iloc[:]
```

```python
fig, ax = plt.subplots(1,1, figsize=(16, 8))
fig.subplots_adjust(hspace=0.4)

plot_1 = total_checkouts.iloc[:].plot( xticks=range(0, 100), xlim=[0, 100], ylim=[0,2000])
plot_1.set_title("Growth of checkouts", fontweight="bold")
plot_1.set(xlabel="Number of people", ylabel="Number of checkouts")
plt.grid()
plt.show()
```
